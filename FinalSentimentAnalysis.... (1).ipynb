# Processing Text

```python
#converting string in lower case
import string
x_train=[comment.lower() for comment in x_train ]
print(x_train[0:11])
```

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
#tokenizing each comment in twitter
token=[]
for comment in x_train:
    newtoken=word_tokenize(comment)
    token.append(newtoken)
print(token[0:20000])
```

```python
table=str.maketrans('','',string.punctuation)
for i in range(len(token)):
    #print(token[i])
    token[i]=[word.translate(table) for word in token[i]]
    token[i]=[word for word in token[i] if word.isalpha()]
print(token[0:11])
```

```python
from nltk.corpus import stopwords
#setting stopwords
stop_words=set(stopwords.words('english'))
#removing stopwords from each comment
for i in range(len(token)):
    token[i]=[w for w in token[i] if not w in stop_words]
print(token[0:11])
```

```python
from nltk.stem.porter import PorterStemmer
porter=PorterStemmer()
stems=[]
for i in range(len(token)):
    stems.append([porter.stem(word) for word in token[i]])
print(stems[0])
```
